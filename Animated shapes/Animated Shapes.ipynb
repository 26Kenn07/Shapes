```python
import turtle
```

```python
def draw_shape(color_shape):
    color_shape = color_shape.lower().split()
    color = None
    shape = None


    color_keywords = {'red', 'green', 'blue', 'yellow', 'orange', 'purple', 'pink'}
    for keyword in color_shape:
        if keyword in color_keywords:
            color = keyword

    shape_keywords = {'circle', 'rectangle', 'triangle', 'pentagon', 'hexagon', 'star', 'diamond', 'heart'}
    for keyword in color_shape:
        if keyword in shape_keywords:
            shape = keyword

    
    if color is None:
        color = 'red'
    if shape is None:
        shape = 'circle'

    
    if color == 'red':
        color_rgb = 'red'
    elif color == 'green':
        color_rgb = 'green'
    elif color == 'blue':
        color_rgb = 'blue'
    elif color == 'yellow':
        color_rgb = 'yellow'
    elif color == 'orange':
        color_rgb = 'orange'
    elif color == 'purple':
        color_rgb = 'purple'
    elif color == 'pink':
        color_rgb = 'pink'

    if shape == 'circle':
        turtle.color(color_rgb)
        turtle.begin_fill()
        turtle.circle(50)
        turtle.end_fill()
    elif shape == 'rectangle':
        turtle.color(color_rgb)
        turtle.begin_fill()
        for _ in range(2):
            turtle.forward(100)
            turtle.right(90)
            turtle.forward(50)
            turtle.right(90)
        turtle.end_fill()
    elif shape == 'triangle':
        turtle.color(color_rgb)
        turtle.begin_fill()
        for _ in range(3):
            turtle.forward(100)
            turtle.left(120)
        turtle.end_fill()
    elif shape == 'pentagon':
        turtle.color(color_rgb)
        turtle.begin_fill()
        for _ in range(5):
            turtle.forward(100)
            turtle.right(72)
        turtle.end_fill()
    elif shape == 'hexagon':
        turtle.color(color_rgb)
        turtle.begin_fill()
        for _ in range(6):
            turtle.forward(100)
            turtle.right(60)
        turtle.end_fill()
    elif shape == 'star':
        turtle.color(color_rgb)
        turtle.begin_fill()
        for _ in range(5):
            turtle.forward(200)
            turtle.right(144)
        turtle.end_fill()
    elif shape == 'diamond':
        turtle.color(color_rgb)
        turtle.begin_fill()
        for _ in range(2):
            turtle.forward(100)
            turtle.right(60)
            turtle.forward(100)
            turtle.right(120)
        turtle.end_fill()
    elif shape == 'heart':
        turtle.color(color_rgb)
        turtle.begin_fill()
        turtle.left(140)
        turtle.forward(180)
        turtle.circle(-90, 200)
        turtle.left(120)
        turtle.circle(-90, 200)
        turtle.forward(180)
        turtle.end_fill()


    turtle.done()
```

```python
input_str = input("Enter the shape and color: ")
draw_shape(input_str)
```
